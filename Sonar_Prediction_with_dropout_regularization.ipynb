```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('sonar_dataset.csv',header=None)# header means
```

```python
df.head()
```

```python
X=df.drop(60,axis='columns')
y=df[60]
```

```python
y.unique()
```

```python
y=y.replace({'R':0,'M':1})
```

```python
y.unique()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0,stratify=y)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
X_train.shape
```

```python
model=keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

Applying Dropout regularization for avoiding overfitting problem


```python
modelb=keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1,activation='sigmoid')

])
modelb.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
modelb.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
modelb.evaluate(X_test,y_test)
```
